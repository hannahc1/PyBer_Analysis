```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

**Line Chart using the MATLAB Approach**
---

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

**To change figure-level attributes (such as axis labels, a title, or a legend) or save the figure as an image, use** *fig = plt.figure().*

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label="Boston")
# Create lables of the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0,45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

**To invert the y-axis to have the months in ascending order, use the gca() method. The gca() method means get current axes. We can chain the gca() method to the invert_yaxis() method by using gca().invert_yaxis()**

**List Comprehension**
---
**new_list = [expression *for* item *in* list *if* conditional]**

```python
fig, ax = plt.subplots()
# Change the color of the markers to sky blue.
# Change the size of the markers to 5 times each data point.
# Make the color 20% transparent.
# Add a black edge color to the circles.
# Make the linewidth of the circles 2 points.
# Add a legend for the city of Boston.
ax.scatter(y_axis, x_axis, 
           s= [i*10 for i in y_axis], 
           color = "skyblue", 
           label = "Boston", 
           alpha = .8, 
           edgecolors = "black", 
           linewidths = (2,))
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
ax.set_xlim(0,50)
ax.invert_yaxis()
```

Pie Chart Using the MATLAB Method
---

```python
plt.subplots(figsize=(8,8))
explode_values = (0,0,0,0,0,0,.5,0,0,0,0,0)
# Assign 12 colors, one for each month.
colors=["slateblue","magenta","lightblue","green","yellowgreen","greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.pie(y_axis, 
        labels=x_axis,
        explode = explode_values, 
        autopct='%.1f%%',
        colors=colors)

plt.show()
```

Pie Chart Using the Object-Oriented Interface
---

```python
fig, ax = plt.subplots(figsize=(8,8))
explode = (0,0,.2,0,0,0,.2,0,0,0,0,0)
colors = ["salmon","darkturquoise","tomato","cadetblue",
          "coral","deepskyblue","orangered","steelblue",
          "lightsalmon","paleturquoise","sienna","aquamarine"]
ax.pie(y_axis,
       labels=x_axis,
       explode=explode,
       autopct = "%.1f%%",
       shadow = True,
       counterclock = False,
       colors = colors,
       startangle = 95)

plt.show()
```
